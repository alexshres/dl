```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

%matplotlib inline
```

```python
# Creating some data
X = np.linspace(-1, 1, 100)
y = 3.0 * X + 2 + np.random.normal(0, 1, size=X.shape)
```

```python
# Defining the model params
w = tf.Variable(initial_value=0.0)
b = tf.Variable(initial_value=0.0)
```

```python
def linear_model(X):
    return w*X + b

def mse(y_pred, y_true):
    return tf.reduce_mean(tf.square(y_pred-y_true))

# tf.square(...) -> computes teh squared differences between y_pred and y_true
# tf.reduce_mean(...) -> calculates average of these squared differences (mse)
```

```python
# Setting up optimizer
optimizer = tf.optimizers.SGD(learning_rate=0.1)

def train(X, w, b, epochs):
    for epoch in range(epochs):
        # tf.GradientTape() -> tracks operations to calculate gradients later
        with tf.GradientTape() as tape:
            y_pred = linear_model(X)
            loss = mse(y_pred, y)

        # Computing gradients of the loss w.r.t. w and b
        gradients = tape.gradient(loss, [w,b])

        # Applying gradients to update vars
        # updating w and b using the computed gradients
        optimizer.apply_gradients(zip(gradients, [w,b]))

        if epoch%10 == 0:
            print(f"Epoch: {epoch}\nLoss: {loss.numpy()}\nw: {w.numpy()}"\
                  f"\nb: {b.numpy()}")
```

```python
train(X, w, b, 100)
```

```python
y_pred = linear_model(X)
plt.figure()
plt.scatter(X, y, label="True Data")
plt.plot(X, y_pred, color="red", label="Model Prediction")

plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Lin. Reg. Model on Training Data")
plt.show()
```

## Now using Keras to build a linear regression model

```python
# Using the same synthetic data as before
X = np.linspace(-1,1,100)
y_true = 3.0 * X + 2.0 + np.random.normal(0, 1, size=X.shape)
```

```python
# Keras expects inputs to be 2d (# of samples, # of features)
X = X.reshape(-1, 1) # -> converting to (100, 1) shape
```

```python
# In Keras, we can define a linear regression model by creating a 
# simple NN dense layer and no activation function

# Define the Keras model
model = Sequential([
    Input(shape=(1,)),             # Since our data only has one feature
    Dense(1, activation='linear')  # 1 unit, linear activation for regression
])

# Dense(1, -> creates a fully connected layer with a single neuron
#          -> which represents y = wx+b
# input_shape=(1,)-> specifies that our input has a single feature
# activation="linear") ensure no non-linear activation is applied
```

```python
# Compiling model using sgd and mse
model.compile(optimizer="sgd", loss="mse")


# Now we TRAIN the model
history = model.fit(X, y_true, epochs=100, verbose=0)
```

```python
final_loss = model.evaluate(X, y_true, verbose=0)
print(f"Final Training Loss (MSE): {final_loss}")
```

```python
def quick_plot(X, y, model):
    y_pred = model.predict(X)

    # Scatterplot
    plt.figure()
    plt.scatter(X,y)

    # Printing Line
    plt.plot(X, y_pred, color="red", label="Model Prediction")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Linear Reg. Model")
    plt.show()
```

```python
quick_plot(X, y_true, model)
```

```python
# Using the history = model.fit(...) to show how the loss decreased

plt.figure()
plt.plot(history.history["loss"], label="Training Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Loss Over Epochs")
plt.legend()
plt.show()
```

```python
# To get the loss after final epoch:
final_loss = history.history["loss"][-1]
print(f"Final training loss: {final_loss}")
```

# Multivariable Linear Regression w/ Regularization

```python
# Create synthetic data

num_samples = 100
num_features = 5

X = np.random.rand(num_samples, num_features)
true_weights = np.array([2.0, -3.5, 1.0, -1.7, 2.1])
true_bias = 4.0

# Adding some noise
y_true = X @ true_weights + true_bias + np.random.normal(0,0.5,size=(num_samples,))
```

```python
# Initializing weights and bias
w = tf.Variable(tf.random.normal([num_features]))
b = tf.Variable(tf.zeros(1))

# Setting regularization factor (lambda)
lambda_reg = 0.01
```

```python
def linear_model(X, w, b):
    return tf.reduce_sum(X*w, axis=1) + b

# tf.reduce_sum(X*w, axis=1) X*w performs element-wise multiplication b/w X and w
# tf.reduce_sum(..., axis=1) sums across each row getting w_1x_1 + ... + w_nx_n
# add b (bias) at end

# mse w/ reg
def mse_w_reg(y_pred, y_true, reg_coeff):
    mse = tf.reduce_mean(tf.square(y_pred-y_true))
    l2_penalty = reg_coeff * tf.reduce_sum(tf.square(w))
    return mse + l2_penalty

# tf.reduce_mean(...) -> computes the mse
# tf.reduce_sum(tf.square(w)) -> sum of squares and sums all of them to get L2 norm of w (vector)
```

```python
# Setting up Optimizer
optimizer = tf.optimizers.SGD(learning_rate=lambda_reg)


# Train the model
def train(X, w, b, y_true, reg_term, epochs):
    for epoch in range(epochs):
        with tf.GradientTape() as tape:
            y_pred = linear_model(X, w, b)
            loss = mse_w_reg(y_pred, y_true, reg_term)

        gradients = tape.gradient(loss, [w,b])
        optimizer.apply_gradients(zip(gradients, [w,b]))

        if epoch%10 == 0:
            print(f"Epoch {epoch}\nLoss: {loss.numpy()}\nWeights: {w.numpy()}"\
                  f"\nBias: {b.numpy()}")
```

```python
train(X, w, b, y_true, lambda_reg, 100)
```

### Now using Keras

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input 
from tensorflow.keras.regularizers import l2 
```

```python
# Using the same data as above
model = Sequential([
    Input(shape=(num_features,)),
    Dense(1, activation="linear", kernel_regularizer=l2(lambda_reg))
])

model.compile(optimizer="sgd", loss="mse")
history = model.fit(X, y_true, epochs=100, verbose=1)
```

```python
final_loss = model.evaluate(X, y_true, verbose=0)
print(f"Final Training Loss (MSE): {final_loss}")
```

```python
plt.figure()
plt.plot(history.history["loss"], label="Training Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss over Epochs")
plt.show()
```
